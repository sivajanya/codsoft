# VISUALISATION

```python
sns.set()
sns.countplot(data=df,x="Pclass",palette="Set3")
plt.title("count of pclass")
plt.show()
```

```python
survived=df["Survived"].value_counts()
plt.figure(figsize=(5,5))
plt.pie(survived,labels=["NOT SURVIVED","SURVIVED"],autopct="%1.1f%%",startangle=90,colors=["lightblue","lightgreen"])
plt.title("survival distribution")
plt.axis("equal")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(),annot=True,cmap="coolwarm")
plt.title("Correlation Matrix",fontsize=20)
```

```python
sns.pairplot(df, hue='Survived', palette='Set1')
plt.suptitle('Pairplot of Titanic Dataset',y=1.0)
plt.show()
```

```python
sns.scatterplot(data=df,x="Age",y="Parch",hue="Survived")
plt.show()
```

```python
sns.scatterplot(x="Age",y="Pclass",data=df,hue="Survived",palette="rocket")
```

```python
sns.histplot(x="Age",data=df,bins=30,kde=True,color="orange")
plt.title("Age Distribution",fontsize=10)
```

```python
sns.set(style="whitegrid")
sns.distplot(df["SibSp"],color="skyblue")
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=df,x="Age",orient="h")
plt.title("boxplot of age",fontsize=20,color="red")
plt.show()
```
